# preprocessing

```python
# converting to lower-case
data["review"] = data["review"].str.lower()
data["sentiment"] = data["sentiment"].str.lower()
```

```python
data.sample(3)
```

```python
# Removing the HTML Tages

import re

def removing_html_tag(text):
    if isinstance(text,str):
        pattern = re.compile("<.*?>")
        return pattern.sub("",text)

    return text
```

```python
data["review"] = data["review"].apply(removing_html_tag)
data["sentiment"] = data["sentiment"].apply(removing_html_tag)
```

```python
data.sample(5)
```

```python
# removing url 

def removing_url(text):
    if isinstance(text,str):
        pattern = re.compile(r"https?://\S+|www\.S+")
        return pattern.sub(r"",text)

    return text
```

```python
data["review"] = data["review"].apply(removing_url)
data["sentiment"] = data["sentiment"].apply(removing_url)
```

```python
data.tail(2)
```

```python
# Removign punctuation from the dataset
import string
pun = string.punctuation

def removing_punctuation(text):
    if isinstance(text,str):
        return text.translate(str.maketrans("","",pun))
    return text
```

```python
data["review"] = data["review"].apply(removing_punctuation)
data["sentiment"] = data["sentiment"].apply(removing_punctuation)
```

```python
# Removign Stop words
import nltk
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

def Removign_stop_words(text):
    if isinstance(text,str):
        return " ".join([i for i in text.split() if i not in stop_words])
    return text
```

```python
data["review"] = data["review"].apply(Removign_stop_words)
```

```python
data.sample(3)
```

```python
# Tokenization
from nltk.tokenize import sent_tokenize,word_tokenize

def word_token(text):
    if isinstance(text,str):
        return word_tokenize(text) 
    raise ValueError("Input must be string")

def sent_token(text):
    if isinstance(text,str):
        return sent_tokenize(text) 
    raise ValueError("Input must be string")
```

```python
data["review"] = data["review"].apply(word_token)
```

```python
data.sample(3)
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer
stems = PorterStemmer()

def stemming(text):
    if isinstance(text,str):
        return text
    else:
        return " ".join([stems.stem(word) for word in text])
```

```python
stemming(["hi"," i", "am","loving","love","scienceing","cricketing"])
```

```python
data["review"] = data["review"].apply(stemming)
```

```python
data.sample(3)
```

```python
# applying the labelencoder an target column
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data["sentiment"] = le.fit_transform(data["sentiment"])
```

# word2vec

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.models import word2vec,keyedvectors
```

```python
story = []

for text in data["review"]:
    raw_sent = sent_tokenize(text)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
len(model.wv.index_to_key)
```

```python
def document_vector(doc):
    doc = [word for word in doc.split() if word in model.wv.index_to_key]
    return np.mean(model.wv[doc],axis = 0)
```

```python
document_vector(data["review"].values[0])
```

```python
from tqdm import tqdm
```

```python
x = []

for doc in tqdm(data["review"].values):
    x.append(document_vector(doc))
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
mnb = GaussianNB()
mnb.fit(X_train,Y_train)
y_pred_m = mnb.predict(X_test)
accuracy_score(Y_test,y_pred_m)
```

```python
rf1 = RandomForestClassifier()

rf1.fit(X_train,Y_train)
y_pred_r = rf1.predict(X_test)
accuracy_score(Y_test,y_pred_r)
```
